# Introduction

Efficient management of delivery routes is a critical challenge in logistics and transportation, especially in urban environments where traffic conditions fluctuate throughout the day. This project focuses on optimizing delivery routes by solving a **Time-Dependent Traveling Salesman Problem (TD-TSP)**, where travel times between locations vary based on traffic flow during rush hours and off-peak periods. 

The core objective is to minimize the total travel duration of a delivery truck visiting multiple customers, starting and ending at a depot. By incorporating time-varying traffic conditions, this model reflects a more realistic approach to route optimization. A weighted undirected graph is used to model the delivery network, with nodes representing customers and edges weighted by time-dependent travel durations. This formulation not only addresses route efficiency but also aims to provide insights into the impact of rush hour traffic on delivery operations.

## Complexity Class$^{[4]}$

**Features:**

### NP

The NP in NP class stands for Non-deterministic Polynomial Time. It is the collection of decision problems that can be solved by a non-deterministic machine in polynomial time. 

**Features**
- The solutions of the NP class are hard to find since they are being solved by a non-deterministic machine but the solutions are easy to verify.

Co-NP stands for the complement of NP Class. It means if the answer to a problem in Co-NP is No, then there is proof that can be checked in polynomial time. 

**Features:**
- If a problem X is in NP, then its complement Xâ€™ is also in CoNP.

### NP-Hard

An NP-hard problem is at least as hard as the hardest problem in NP and it is a class of problems such that every problem in NP reduces to NP-hard.

**Features:**
- All NP-hard problems are not in NP.
- It takes a long time to check them. This means if a solution for an NP-hard problem is given then it takes a long time to check whether it is right or not.

### NP-Complete

A problem is NP-complete if it is both NP and NP-hard. NP-complete problems are the hard problems in NP.

**Features:**
- NP-complete problems are special as any problem in NP class can be transformed or reduced into NP-complete problems in polynomial time.

## Travelling Salesman Problem

### What is TSP?

Mathematicians and computer scientists have paid close attention to the traveling salesman problem (TSP) because it is both straightforward to define and challenging to solve. The problem can be expressed simply as follows: what is the least expensive path a traveling salesperson can take if he wants to visit each of a list of m towns exactly once (where the cost of going from $city_i$ to $city_j$ is $c_{ij}$) and then return to his home city?

### Importance of TSP

The TSP is significant because it exemplifies a broader category of issues referred to as combinatorial optimization problems. The TSP issue (*decision problem*) falls within the category of **NP-complete** problems$^{[7]}$, however its *optimization problem* is considered to be a part of the **NP-Hard** class.$^{[8]}$ 

### Symmetric and Asymmetric TSP$^{[9]}$

In the symmetric TSP, the distance between two cities is the same in each opposite direction, forming an undirected graph. This symmetry halves the number of possible solutions.

In the asymmetric TSP, paths may not exist in both directions or the distances might be different, forming a directed graph. Traffic congestion, one-way streets, and airfares for cities with different departure and arrival fees are real-world considerations that could yield a TSP problem in asymmetric form.

## Vehicle Routing Problem

### What is VRP? $^{[11]}$

The VRP can be represented as the following graph-theoretic problem. 

Let G = (V, A) be a complete graph where V = {0, 1, . . . ,n} is the vertex set and A is the arc set. 

Vertices j = 1, . . . , n correspond to the customers, each with a known non-negative demand, $d_j$, whereas vertex 0 corresponds to the depot. 

A non-negative cost, $c_{ij}$, is associated with each arc $(i,j) \in A$ and represents the cost of traveling from vertex i to vertex j. 

- If the cost values satisfy $c_{ij} = c_{ji} \quad \forall  (i,j) \in V$, then the problem is said to be a **symmetric VRP**; 
- otherwise, it is called an **asymmetric VRP**.

In several practical cases the cost matrix satisfies the *triangle inequality*, such that $c_{ik} + c_{kj} \geq c_{ij} \quad for \ any \quad i, j, k \in V$

### Importance of VRP

Due to their significance in numerous application domains, such as manufacturing, logistics, communications, transportation, military and relief systems, as well as their high scientific interest as challenging combinatorial optimization problems, vehicle routing problems have been the focus of extensive research for more than 50 years$^{[13]}$. The first document$^{[10]}$ in the VRP literature, examined a TSP on a somewhat large scale, and suggested a solution. Numerous further TSP articles were published after that investigation which demonstrated that TSP is a particular instance of VRP.

VRP *optimization problem* is considered **NP-Hard**$^{[12,14]}$. However, if we reduce the VRP optimization problem to a *decision problem*, it will be considered as **NP-Complete**.

The VRP constitutes a generalization of the TSP$^{[15]}$ which consists of determining the shortest circuit or cycle passing through each of the n points once and only once. Though TSP and the VRP are both NP-hard (optimization problems), from a practical point of view, VRPs are in general much more difficult to solve than TSPs of the same size.

## Choosing TD-TSP instead of TD-VRP

### What is TD-TSP and TD-VRP?$^{[17]}$

Time dependent vehicle routing problem (TD-VRP) is defined as follows:

A vehicle fleet of fixed capacities has to serve customers of fixed demands from a central depot. Customers must be assigned to vehicles and the vehicles routed so that the total time spent on the routes is minimized. The travel time between two customers or between a customer and a depot *depends on the distance between the points and the time of the day*.

Time dependent salesman problem (TD-TSP) is a **special case** of the TD-VRP in which only one vehicle of infinite capacity is available.

### Understanding our problem

We are:
- Considering 1 truck and 1 depot
- **NOT** considering truck capacity or delivery load balancing
- Primarily focusing on optimizing *route duration* with *varying travel times* due to traffic (central feature of TD-TSP)

Therefore, our problem is more aligned with **TD-TSP** rather than **TD-VRP**.

# Decision Variables

$
x_{ij} = 
\begin{cases} 
1 & \text{if truck travels from node } i \text{ to node } j \\
0 & \text{otherwise} 
\end{cases}
$

- $x_{ij}$: A binary variable that indicates whether the truck travels from node i to node j

$
t_{ij}(t_j) = 
\begin{cases} 
t_{ij}^{\text{low}} & \text{if } t_i \in \text{off-peak hours}, \\
t_{ij}^{\text{high}} & \text{if } t_i \in \text{peak hours}.
\end{cases}
$

- $t_{ij}(t_i)$: Continuous variable representing the travel time between nodes i and j, which depends on the time the truck arrives at node i
- $t_j$: Continuous decision variable representing the time the truck arrives at node j

# Constraints

$
\sum_{j=1, \, j \neq i}^{n} x_{ij} = 1 \quad \forall i = 1, \dots, n \quad \text{(the truck leaves node } i \text{ once)}
$

$
\sum_{i=1, \, i \neq j}^{n} x_{ij} = 1 \quad \forall j = 1, \dots, n \quad \text{(the truck arrives at node } j \text{ once)}
$

$
t_j \geq 0 \quad \forall j \quad \text{(the time the truck arrives at node j)}
$

$
t_0 \geq 0 \quad \text{(the time the truck departs from the depot (node 0))}
$

$
t_j \geq t_i + t_{ij}(t_i) \cdot x_{ij} \quad \forall i, j, \quad i \neq j \quad \text {(The arrival time at node j should depend on the arrival time at node i and the travel time }t_{ij}(t_i){ )}
$

$
u_i - u_j + (n - 1) \cdot x_{ij} \leq n - 2 \quad \forall i \neq j \quad \text {(MTZ Subtour Elimination Constraints)}
$

- $u_i$ is an auxiliary variable representing the position of node i in the tour (for all nodes ð‘– â‰  0, where node 0 is the depot).
- $u_1, ..., u_n$ are continuous variables constrained by 2 â‰¤ $u_i$ â‰¤ n.
- This constraint ensures that nodes are visited in the correct order and prevents subtours.

# Fitness Function

$
T = t_0 + \sum_{i=0}^{n} \sum_{j=1}^{n} t_{ij}(t_j) \cdot x_{ij}
$

# Objective Function

$
Minimize \quad T = t_0 + \sum_{i=0}^{n} \sum_{j=1}^{n} t_{ij}(t_j) \cdot x_{ij}
$

# Building the Graph

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
random.seed(3)
```

```python
NUM_NODES = 24
RUSH_HOURS = [(7, 9), (16, 19)]  # Representing 7h-9h and 16h-19h
```

```python
# Function to determine if a given hour is within rush hours
def is_rush_hour(hour):
    return any(start <= hour < end for start, end in RUSH_HOURS)
```

```python
G = nx.complete_graph(NUM_NODES)
```

```python
for (i, j) in G.edges:
    base_time = random.randint(60, 180)  # Random base travel time in minutes
    off_peak_time = base_time
    rush_hour_time = base_time + (random.randint(20,30) * base_time) / 100
    G.edges[i, j]['off_peak_time'] = off_peak_time
    G.edges[i, j]['rush_hour_time'] = rush_hour_time
```

```python
plt.figure(figsize=(15, 15))
pos = nx.spring_layout(G, seed=42)  # Position the nodes using a spring layout
nx.draw(G, pos, with_labels=True, node_size=700, node_color="skyblue", font_size=10, font_weight="bold")

# Create labels for each edge showing the off-peak travel time
edge_labels = {(i, j): f"{G.edges[i, j]['off_peak_time']} / {G.edges[i, j]['rush_hour_time']}" for i, j in G.edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

plt.title("Complete Graph with 24 Nodes\nEdge Labels Show - Off-Peak / Rush Hour Travel Times")
plt.show()
```

```python
# Function to get travel time based on hour of the day
def get_travel_time(i, j, hour):
    """Returns travel time for edge (i, j) based on the hour."""
    edge_data = G.get_edge_data(i, j)
    if edge_data is None:
        raise ValueError(f"No edge exists between nodes {i} and {j}")
    
    if is_rush_hour(hour):
        return edge_data['rush_hour_time']
    else:
        return edge_data['off_peak_time']
```

# Genetic Algorithm$^{[22]}$

## Overview

Genetic Algorithms (GAs) are adaptive heuristic search algorithms that form part of the evolutionary algorithms family. They are inspired by natural selection and genetic processes found in biological systems. GAs are effective for solving optimization and search problems by emulating the principle of "survival of the fittest."

---

## Core Components of Genetic Algorithms

1. **Population**: A group of potential solutions (individuals) to the problem.
   - Each individual represents a candidate solution and is encoded as a vector of values (often binary, integer, or float).
   - These values are analogous to chromosomes in biology.

2. **Fitness Score**: A metric indicating how well an individual solves the problem.
   - Individuals with higher fitness scores have a better chance of being selected for reproduction, leading to improved solutions over generations.

3. **Generations**: Each new generation of individuals is created from the previous one.
   - Fitter individuals have a higher likelihood of producing offspring, which inherit characteristics that may improve overall solution quality.
   - Over time, the population "converges" on better solutions.

---

## Foundations of Genetic Algorithms

The genetic algorithm process mimics natural selection:
- **Selection**: Only the fittest individuals survive and reproduce.
- **Propagation**: Genetic information from the most fit individuals spreads throughout the population.
- **Evolutionary Improvement**: Successive generations exhibit increasingly optimal traits suited for solving the problem.

---

## Search Space Representation

The **search space** is the domain of possible solutions for a given problem. Each individual represents a unique point in this search space:
- Each individual is encoded as a finite-length vector (chromosome) composed of multiple components (genes).
- Genes represent variable components that contribute to the individualâ€™s solution, analogous to biological genes.

---

## Operators in Genetic Algorithms

GAs use three primary operators to evolve the population and seek better solutions:

1. **Selection Operator**: 
   - Favors individuals with higher fitness scores.
   - Ensures that "stronger" genes are passed on to the next generation, improving the population's overall fitness.

2. **Crossover Operator**:
   - Simulates mating between individuals.
   - Two individuals are selected, and crossover points are chosen randomly to swap segments of their genes, producing offspring with mixed traits.

3. **Mutation Operator**:
   - Introduces random gene alterations in offspring to maintain diversity in the population.
   - Helps prevent "premature convergence" where the algorithm might otherwise get stuck in suboptimal solutions.

```python
# Function to calculate the fitness of a route
def calculate_route_time(route, start_hour):
    """Calculate the total travel time for a route considering time-dependent traffic."""
    total_time = 0
    current_hour = start_hour
    for i in range(len(route) - 1):
        u, v = route[i], route[i + 1]
        travel_time = get_travel_time(u, v, current_hour)
        total_time += travel_time
        current_hour = (current_hour + (travel_time // 60)) % 24  # Update hour, wrapping to 24-hour format
    
    # Return back to the depot (node 0)
    if route[-1] == 0:
        travel_time = 0  # No time to return to itself
    else:
        travel_time = get_travel_time(route[-1], 0, current_hour)  # Get travel time from the last node to depot
    total_time += travel_time
    return total_time
```

```python
# Initialize population
def initialize_population():
    population = []
    for _ in range(POP_SIZE):
        route = list(range(1, NUM_NODES))  # Start from 1 (exclude depot node 0)
        random.shuffle(route)
        route = [0] + route + [0]  # Complete the route to form a Hamiltonian circuit
        population.append(route)
    return population
```

```python
# Selection using tournament selection
def tournament_selection(population, fitnesses):
    tournament = random.sample(list(zip(population, fitnesses)), TOURNAMENT_SIZE)
    tournament.sort(key=lambda x: x[1])
    return tournament[0][0]
```

## Some Types of Selection Process

1. **Roullette Wheel Selection**: In the roulette wheel selection, the probability of choosing an individual for breeding of the next generation is proportional to its fitness, the better the fitness is, the higher chance for that individual to be chosen. Choosing individuals can be depicted as spinning a roulette that has as many pockets as there are individuals in the current generation, with sizes depending on their probability.
2. **Rank Selection**: In rank selection, the selection probability does not depend directly on the fitness, but on the fitness rank of an individual within the population. This puts large fitness differences into perspective; moreover, the exact fitness values themselves do not have to be available, but only a sorting of the individuals according to quality.
3. **Tournament Selection**: Tournament selection is a method of choosing the individual from the set of individuals. The winner of each tournament is selected to perform crossover.
4. **Elitist Selection**: Often to get better results, strategies with partial reproduction are used. One of them is elitism, in which a small portion of the best individuals from the last generation is carried over (without any changes) to the next one.

### Why did we choose Tournament Selection?

Tournament selection is effective for balancing solution diversity and efficiency. In each tournament, a subset of individuals competes, and the best individual in this subset is chosen for crossover. This method encourages diversity by giving weaker solutions a chance to survive while maintaining a competitive selection process, reducing the risk of premature convergence.

### Why not Roulette-Wheel or Rank Selection?

Roulette-Wheel Selection assigns higher probabilities to fitter individuals, but it can lead to *excessive selection pressure* in some cases, causing premature convergence, especially in cases with highly variable fitness levels.

Rank Selection is more stable but is *computationally expensive*, especially with larger populations, as it requires sorting each generation. 

Tournament selection provides a balance between selective pressure and computation, which is ideal for complex problems like TD-TSP.

```python
# Crossover operation
def ordered_crossover(parent1, parent2):
    size = len(parent1)
    start, end = sorted(random.sample(range(1, size - 1), 2))
    child = [None] * size
    child[start:end] = parent1[start:end]
    
    # Fill in the rest with nodes from the second parent
    current_index = 0
    for node in parent2:
        if node not in child:
            # Find the next None position in the child
            while current_index < size and child[current_index] is not None:
                current_index += 1
            if current_index < size:  # Ensure we don't go out of bounds
                child[current_index] = node

    # Ensure the child starts and ends with the depot (node 0)
    child[0] = 0
    child[-1] = 0

    # Fill any remaining None values with valid nodes
    for i in range(size):
        if child[i] is None:
            # Find a valid node not already in the child
            for node in parent2:                
                if node not in child:
                    child[i] = node
                    break
    return child
```

```python
# Mutation operation
def mutate(route):
    if random.random() < MUTATION_RATE:
        i, j = sorted(random.sample(range(1, len(route) - 1), 2))
        route[i], route[j] = route[j], route[i]
```

```python
# Main Genetic Algorithm function
def genetic_algorithm(start_hour=0, POP_SIZE=POP_SIZE, NUM_GENERATIONS=NUM_GENERATIONS, MUTATION_RATE=MUTATION_RATE, 
                      TOURNAMENT_SIZE=TOURNAMENT_SIZE, ELITISM_RATE=ELITISM_RATE, STAGNATION_LIMIT=STAGNATION_LIMIT):
    population = initialize_population()
    fitness_history = []
    best_route, best_fitness = None, float('inf')
    stagnation_counter = 0

    for generation in range(NUM_GENERATIONS):
        # Calculate fitness for each route
        fitnesses = [calculate_route_time(route, start_hour) for route in population]
        
        # Log best fitness of the current generation
        best_gen_fitness = min(fitnesses)
        if best_gen_fitness < best_fitness:
            best_fitness = best_gen_fitness
            best_route = population[fitnesses.index(best_fitness)]
            stagnation_counter = 0
        else:
            stagnation_counter += 1
        
        ## logging.info(f"Generation {generation + 1}: Best fitness = {best_gen_fitness}")
        
        # Stopping condition for stagnation
        if stagnation_counter >= STAGNATION_LIMIT:
            break

        # Selection and Crossover
        new_population = []
        elite_size = int(ELITISM_RATE * POP_SIZE)
        elites = [population[i] for i in np.argsort(fitnesses)[:elite_size]]
        new_population.extend(elites)

        while len(new_population) < POP_SIZE:
            parent1 = tournament_selection(population, fitnesses)
            parent2 = tournament_selection(population, fitnesses)
            child = ordered_crossover(parent1, parent2)
            mutate(child)
            new_population.append(child)
        
        population = new_population

    return best_route, best_fitness
```

 [Ant colony](https://www.sciencedirect.com/topics/agricultural-and-biological-sciences/ant-colonies "Learn more about Ant colony from ScienceDirect's AI-generated Topic Pages")Â optimization (ACO) is aÂ [metaheuristic optimization](https://www.sciencedirect.com/topics/engineering/metaheuristic-optimization "Learn more about metaheuristic optimization from ScienceDirect's AI-generated Topic Pages")Â technique based on theÂ [behavior](https://www.sciencedirect.com/topics/neuroscience/behavior-neuroscience "Learn more about behavior from ScienceDirect's AI-generated Topic Pages")Â of ants. $[25]$

**Ant Colony Optimization (ACO)**Â is a nature-inspired algorithm that simulates the behavior of ants searching for food to solve complex optimization problems, particularly those that involve finding the shortest or most efficient path, such as theÂ **Traveling Salesman Problem (TSP)**Â orÂ **route optimization**.

In nature, ants leave a trail ofÂ **pheromones**Â on the ground as they explore paths. Other ants are more likely to follow paths with stronger pheromone trails, which leads them to the shortest routes over time. *( the more ants that follow a trail, the stronger the pheromone trail becomes.)* ACO algorithms mimic this behavior by having virtual "ants" explore potential solutions and update "pheromone" trails to guide future solutions.$[26]$

## Core Components of the Ant Colony Optimization :

These parameters define the behavior and balance of the ACO algorithm: 

-**NUM_ANTS**: The number of ants exploring paths in each iteration.

- **PHEROMONE_WEIGHT**Â (`Î±`): Determines how much influence the pheromone level has on an antâ€™s decision.  
- **HEURISTIC_WEIGHT**Â (`Î²`): Determines the influence of heuristic information (shorter paths are favored more with higher values).  
- **Q**: A scaling factor for the amount of pheromone deposited by ants.

## Foundations of ACO Algorithm :  


 $[27]$  


####  **Step 1: Initial Exploration**
Ants leave the **nest (N)** and explore paths to the **food source (F)** randomly, as no pheromone trails exist yet. They deposit low levels of pheromone on the paths they traverse, spreading trails across multiple routes. The goal is to explore all possible routes for future reference.
####  **Step 2: Path Reinforcement and Return**
After reaching the food source, ants return to the nest, reinforcing the paths they took by depositing more pheromones. Shorter or faster paths accumulate more pheromones since they are completed more quickly, attracting more ants. However, some ants still explore other paths.
####  **Step 3: Convergence on Optimal Path**
Over time, the most efficient paths build up the highest pheromone levels, while less efficient paths lose pheromone through evaporation. This leads most ants to converge on the optimal path, with pheromone evaporation ensuring only the best routes remain strong.    


### ->**Exploration & Exploitation in ACO** :  

- **Exploration**: This is the phase where ants search different paths to discover all possible routes between the nest and the food source. In the image, **Step 1** represents exploration, where ants choose paths randomly without any prior knowledge. Exploration is essential to ensure that ants donâ€™t miss potentially good routes by only following initial trails.

- **Exploitation**: This phase leverages the pheromone trails laid down during exploration. As certain paths accumulate higher pheromone concentrations (because they are shorter or more efficient), ants increasingly follow these paths. **Step 2** and **Step 3** in the image illustrate exploitation, where ants follow the reinforced paths, converging on the best path over time.  

  **Balancing Exploration and Exploitation**:
  - In ACO, itâ€™s crucial to balance exploration and exploitation:
  - **Too much exploration** means ants keep trying random paths without converging on the best solution.
  - **Too much exploitation** leads to ants quickly converging on a path that might not be optimal (a local optimum).

```python
# ACO Parameters
NUM_ANTS = 50               # Number of ants per iteration
NUM_ITERATIONS = 200        # Number of iterations
EVAPORATION_RATE = 0.5      # Pheromone evaporation rate
PHEROMONE_WEIGHT = 1.0      # Weight of pheromone influence (alpha)
HEURISTIC_WEIGHT = 2.0      # Weight of heuristic influence (beta)
Q = 1000                    # Constant for pheromone deposit
```

```python
# Initialize pheromone levels
pheromone_matrix = np.ones((NUM_NODES, NUM_NODES))
```

```python
def heuristic(i, j, current_time):
    """Heuristic function (Î·), inverse of travel time."""
    travel_time = get_travel_time(i, j, current_time)
    return 1.0 / (travel_time + 1e-5)  # The small value prevents division by zero
```

```python
def construct_solution(start_hour):
    """Constructs a route for an ant based on current pheromones and heuristics, considering time dependency."""
    unvisited = set(range(1, NUM_NODES))
    route = [0]  # Start at depot
    current_node = 0
    current_time = start_hour  # Initialize with the start hour
    
    while unvisited:
        probabilities = []
        total_pheromone_heuristic = 0
        
        for next_node in unvisited:
            travel_time = get_travel_time(current_node, next_node, current_time)
            pheromone = pheromone_matrix[current_node, next_node] ** PHEROMONE_WEIGHT
            heuristic_value = heuristic(current_node, next_node, current_time) ** HEURISTIC_WEIGHT
            probabilities.append(pheromone * heuristic_value)
            total_pheromone_heuristic += pheromone * heuristic_value
        
        # Normalize probabilities
        probabilities = [p / total_pheromone_heuristic for p in probabilities]
        next_node = random.choices(list(unvisited), weights=probabilities)[0]
        
        # Update current time with travel time to next node
        current_time = (current_time + get_travel_time(current_node, next_node, current_time)) % 24
        
        route.append(next_node)
        unvisited.remove(next_node)
        current_node = next_node
    
    route.append(0)  # Return to depot
    return route
```

```python
def update_pheromones(routes, fitnesses):
    global pheromone_matrix
    pheromone_matrix *= (1 - EVAPORATION_RATE)  # Apply evaporation

    # Update pheromones based on all routes
    for route, fitness in zip(routes, fitnesses):
        pheromone_deposit = 1 / fitness
        for i in range(len(route) - 1):
            u, v = route[i], route[i + 1]
            pheromone_matrix[u, v] += Q * pheromone_deposit
            pheromone_matrix[v, u] += Q * pheromone_deposit  # Symmetric update for undirected graph
```

```python
def ant_colony_optimization(start_hour=0, NUM_ANTS=NUM_ANTS, NUM_ITERATIONS=NUM_ITERATIONS, EVAPORATION_RATE=EVAPORATION_RATE, 
                            PHEROMONE_WEIGHT=PHEROMONE_WEIGHT, HEURISTIC_WEIGHT=HEURISTIC_WEIGHT, Q=Q):
    best_route, best_fitness = None, float('inf')
    for iteration in range(NUM_ITERATIONS):
        routes = []
        fitnesses = []
        
        for _ in range(NUM_ANTS):
            route = construct_solution(start_hour)
            travel_time = calculate_route_time(route, start_hour)
            routes.append(route)
            fitnesses.append(travel_time)
        
        # Update global best solution
        iteration_best_fitness = min(fitnesses)
        if iteration_best_fitness < best_fitness:
            best_fitness = iteration_best_fitness
            best_route = routes[fitnesses.index(best_fitness)]
        
        update_pheromones(routes, fitnesses)
        
        ## logging.info(f"Iteration {iteration + 1}: Best fitness = {iteration_best_fitness}")
    
    return best_route, best_fitness
```
